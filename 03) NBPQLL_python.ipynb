```python
pysql = lambda q: pdsql.sqldf(q, globals())
```

```python
str1 = """
  select *, 
    case when length(SS7) = 5 then 'A00'|| SS7 
         when length(SS7) = 6 then 'A0'|| SS7 
          else 'A' || SS7 end as SSN7 
  from df a;
  """
```

```python
df = pysql(str1)
df.head()
```
